```python
#Baixei as bibliotecas necessárias
#Lê e acessa arquivos
import pandas as pd
#Faz e facilita operações matemáticas
import numpy as np
#Auxilia com a criação de gráficos
import matplotlib.pyplot as plt

dataset = pd.read_excel(r"C:\Users\henrique_dahmer\Downloads\producao_motores_weg_iot_10000.xlsx")
display(dataset)
```

```python
#Renomeia as colunas para o padrão de utilização da WEG
dataset.columns = ['data_hora', 'setor', 'maquina', 'cod_motor', 'sensor', 'temperatura', 'vibracao', 'corrente', 'tensao', 'pressao', 'umidade', 'status', 'cod_falha', 'operador', 'turno', 'consumo_kwh']

#Exibe a mudança executada
dataset.head()
```

```python
#Motor
#remover espaços
dataset['cod_motor'] = dataset['cod_motor'].str.strip()

#converter para letras maiúsculas
dataset['cod_motor'] = dataset['cod_motor'].str.upper()

#eliminar registros duplicados
dataset = dataset.drop_duplicates(subset=['cod_motor'], keep='first')

#Resultado:
dataset.head()
```

```python
#Operadores
#primeira letra maiúscula
dataset['operador'] = dataset['operador'].str.title()

#remoção de espaços extras
dataset['operador'] = dataset['operador'].str.strip()

#Resultado: 
dataset.head()
```

```python
#Status
#Padronizar para apenas:
# Aprovado
# Reprovado 
#Cria o dicionário com as regras de substituição
substituicoes = {
    #Tudo que for variação de "OK" ou "Aprovado" vira 'Aprovado'
    'OK': 'Aprovado',
    'Ok': 'Aprovado',
    'ok': 'Aprovado',
    'APROVADO': 'Aprovado',
    'aprovado': 'Aprovado',
    
    #Tudo que for variação de "Reprovado" ou "Falha" vira 'Reprovado'
    'REPROVADO': 'Reprovado',
    'Falha': 'Reprovado',
    'FALHA': 'Reprovado'
}

#Aplica as substituições na coluna 'status'
dataset['status'] = dataset['status'].replace(substituicoes)

#Contagem final
dataset['status'].value_counts()
```

```python
#Conta tudo, com a intenção de ver os nomes
dataset['turno'].value_counts()
```

```python
#Turnos
#Padronizar para:
# 1º Turno
# 2º Turno
# 3º Turno

#Conta tudo, com a intenção de ver os nomes
dataset['turno'].value_counts()


turnos = {
    'Primeiro': '1º Turno',
    '1': '1º Turno',
    'Segundo': '2º Turno',
    '2': '2º Turno',
    'Terceiro': '3º Turno',
    '3': '3º Turno',
    'Madrugada': '3º Turno'
}

#Substitui os nomes para o padrão definido
dataset['turno'] = dataset['turno'].replace(turnos)

#Contagem final
dataset['turno'].value_counts()
```

```python
#Código de Falha
#Substituir:
#• NULL
#• células vazias
# por:
# SEM_FALHA

# Preenche valores nulos/vazios (NaN/NULL) por 'SEM_FALHA'
dataset['cod_falha'] = dataset['cod_falha'].fillna('SEM_FALHA')

# Se no seu arquivo houver traços como '---' ou palavras vazias, substitui por 'SEM_FALHA' também
dataset['cod_falha'] = dataset['cod_falha'].replace({'---': 'SEM_FALHA', '': 'SEM_FALHA'})

# Confere o resultado
dataset['cod_falha'].value_counts()
```

```python
#Temperatura
#Substituir:
# valores negativos;
# valor 999;
# valores nulos.

# Cálculo da mediana com dados válidos
filtro_validos = (dataset['temperatura'] != 999) & (dataset['temperatura'] >= 0)
mediana_temperatura = dataset[filtro_validos]['temperatura'].median()
print(f"A mediana correta é: {mediana_temperatura}")

# Substituindo os negativos e o 999 por NaN
dataset.loc[(dataset['temperatura'] == 999) | (dataset['temperatura'] < 0), 'temperatura'] = np.nan

# Preenchendo todos os vazios
dataset['temperatura'] = dataset['temperatura'].fillna(mediana_temperatura)

# Resultado final:
dataset['temperatura'].value_counts()
```

```python
#Vibração
#Substituir:
#    valores negativos e 
#    acima do limite operacional 
#    pela 
#    mediana.


# Defini aqui qual é o limite operacional de vibração da máquina
limite_operacional = 20.0

# Cálculo da mediana com dados válidos
filtro_validos = (dataset['vibracao'] <= limite_operacional) & (dataset['vibracao'] >= 0)
mediana_vibracao = dataset[filtro_validos]['vibracao'].median()
print(f"A mediana correta é: {mediana_vibracao}")

# Substituindo os valores negativos e os acima do limite por NaN (Nulos)
dataset.loc[(dataset['vibracao'] > limite_operacional) | (dataset['vibracao'] < 0), 'vibracao'] = np.nan

# Preenchendo todos com a mediana
dataset['vibracao'] = dataset['vibracao'].fillna(mediana_vibracao)

# Resultado final
dataset['vibracao'].describe()
```

```python
#Corrente
#Substituir:   
# valores iguais a zero
# pela
# mediana

#Cálculo da mediana com os valores válidos
filtro_validos = (dataset['corrente'] != 0)
mediana_corrente = dataset[filtro_validos]['corrente'].median()
print(f"A mediana correta é: {mediana_corrente}")

# Preenchendo todos com a mediana
dataset['corrente'] = dataset['corrente'].replace(0.00, mediana_corrente)

# Resultado final
dataset['corrente'].value_counts()
```

```python
#Tensão
#Corrigir valores fora da faixa operacional.

# Definindo apenas o limite operacional
limite_operacional = 245.0

# Cálculo da mediana com dados válidos (maior ou igual a 0 E menor ou igual ao limite)
filtro_validos = (dataset['tensao'] <= limite_operacional) & (dataset['tensao'] >= 0)
mediana_tensao = dataset[filtro_validos]['tensao'].median()
print(f"A mediana correta é: {mediana_tensao}")

# Substituindo os valores negativos e os acima do limite por NaN
dataset.loc[(dataset['tensao'] > limite_operacional) | (dataset['tensao'] < 0), 'tensao'] = np.nan

# Preenchendo todos os nulos com a mediana
dataset['tensao'] = dataset['tensao'].fillna(mediana_tensao)

# Resultado final para conferência
dataset['tensao'].describe()
```

```python
#Pressão
#Substituir valores acima do limite permitido utilizando a mediana.

# Definindo apenas o limite operacional
limite_operacional = 15.0

# Cálculo da mediana com dados válidos 
filtro_validos = (dataset['pressao'] <= limite_operacional) & (dataset['pressao'] >= 0)
mediana_pressao = dataset[filtro_validos]['pressao'].median()
print(f"A mediana correta é: {mediana_pressao}")

# Substituindo os valores negativos e os acima do limite por NaN
dataset.loc[(dataset['pressao'] > limite_operacional) | (dataset['pressao'] < 0), 'pressao'] = np.nan

# Preenchendo todos os nulos com a mediana
dataset['pressao'] = dataset['pressao'].fillna(mediana_pressao)

# Resultado final
dataset['pressao'].describe()
```

```python
#Consumo de Energia
#Tratar valores muito elevados utilizando a mediana.

# Definindo apenas o limite operacional
limite_operacional = 250.0

# Cálculo da mediana com dados válidos 
filtro_validos = (dataset['consumo_kwh'] <= limite_operacional) & (dataset['consumo_kwh'] >= 0)
mediana_consumo_kwh = dataset[filtro_validos]['consumo_kwh'].median()
print(f"A mediana correta é: {mediana_consumo_kwh}")

# Substituindo os valores negativos e os acima do limite por NaN
dataset.loc[(dataset['consumo_kwh'] > limite_operacional) | (dataset['consumo_kwh'] < 0), 'consumo_kwh'] = np.nan

# Preenchendo todos os nulos com a mediana
dataset['consumo_kwh'] = dataset['consumo_kwh'].fillna(mediana_consumo_kwh)

# Resultado final
dataset['consumo_kwh'].describe()
```
