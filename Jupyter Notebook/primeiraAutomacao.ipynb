```python
import pandas as pd
import time
tabela = pd.read_excel(r"C:\Users\henrique_dahmer\Downloads\Vendas_WEG_Pecas_Industriais.xlsx")
display(tabela)

time.sleep(3)

tabela['Valor Final'] = (
    tabela['Valor Final']
    .str.replace("R$", "", regex = False)
    .str.replace(".", "", regex = False)
    .str.replace(",", ".", regex = False)
    .str.strip()
)

tabela['Valor Final'] = tabela['Valor Final'].astype(float)

faturamento = tabela['Valor Final'].sum();
print("R$ ", faturamento)
```
